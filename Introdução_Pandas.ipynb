# **Introdução ao Pandas**

### Importando o Pandas

```python
import pandas as pd
```

### Importando dataframe apartir de uma URL

```python
# Link do dataframe no formato .csv
url = 'https://raw.githubusercontent.com/andersonnb6/Databases/main/mtcars.csv'

# Convertendo .csv em dataframe pandas
df = pd.read_csv(url)
```

### Visualizando dataframe

```python
# Mostrando todo o dataframe
display(df)
```

```python
# Mostrando 5 primeiras linhas
df.head(5)
```

```python
# Mostrando total de linhas e colunas, respectivamente
df.shape
```

### Descrevendo resumo dos dados

```python
# Resumindo todas variáveis numéricas
df.describe()
```

```python
# Resumindo uma variável específica (mpg)
df.mpg.describe()
```

# Edição de dataframes

```python
# Convertendo uma coluna em SÉRIE a pandas
modelos = df['model']
modelos.head()
```

```python
# Reduzindo dataframe em colunas específicas
df_potencia = df[['model','hp']]
df_potencia.head()
```

```python
# Pegar dados uma linha específica pelo índice
df.loc[0]
```

```python
# Pegar dados de um intervalo de linhas específicos
df.loc[0:3]
```

```python
# Pegar linhas de acordo com uma condição
df.loc[df['mpg'] > 30]
```

```python
# Filtrando linhas e colunas ao mesmo tempo
# df.loc[linha, colunas]
df.loc[df['mpg'] > 30, ['model', 'mpg']]
```

```python
# Pegar um valor específico 
# df.loc[linha, colunas]
df.loc[2, 'model']
```

```python
# Criar uma coluna a partir de outra pré-existente
df['dobro_hp'] = df['hp']*2
df.head(3)
```

```python
# Criando coluna onde todos as linhas são zero (0)
# O sinal : indica todas as linhas
df.loc[ :, 'nova_coluna'] = 0
df.head(3)
```

```python
# Excluir colunas. Quando axis = 1 indica que estamos removendo colunas.
df = df.drop('nova_coluna', axis = 1)
df.head(3)
```

```python
# Excluir linha 2. Quando axis = 0 indica que estamos removendo linhas.
df = df.drop(2, axis = 0)
df.head(3)
```

```python
# Deletar linhas com TODOS os valores vazios
df = df.dropna(how = 'all')
df.head(3)
```

```python
# Deletar colunas com TODOS os valores vazios
df = df.dropna(how = 'all', axis = 1)
df.head(3)
```

```python
# Deletar linhas com PELO AO MENOS 1 valor vazio
df = df.dropna()
df.head(3)
```
